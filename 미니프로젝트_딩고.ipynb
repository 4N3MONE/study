```python
import datetime
import requests 
from selenium import webdriver 
import time 
import re 
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
keyword = "딩고" 
url = "https://www.instagram.com"
url1 = "https://www.instagram.com/explore/tags/{}/".format(keyword) 

filename = 'c:\\tmp\\크롤링\\인스타크롤링_'+keyword+'.csv'
print("="*70)
print("인스타그램 딩고 크롤러 입니다.")
searchcount = int(input("검색할 건수를 입력해 주세요."))
instaid = input("인스타그램에 로그인할 때 필요한 아이디를 입력해주세요.")
instapw = input("인스타그램에 로그인할 때 필요한 비밀번호를 입력해주세요.")
print("="*70)


instagram_tags = [] 
path = "c:\\tmp\\chromedriver.exe"
driver = webdriver.Chrome(path)

driver.get(url) 
time.sleep(3) 

#로그인
driver.find_elements_by_name('username')[0].send_keys(instaid)
driver.find_elements_by_name('password')[0].send_keys(instapw)
driver.find_element_by_xpath('//*[@id="loginForm"]/div/div[3]/button').click()
time.sleep(3)

#팝업창 처리
driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/div/div/section/div/button').click()
time.sleep(3)
driver.find_element_by_xpath('/html/body/div[4]/div/div/div/div[3]/button[2]').click()

#검색어 url 진입
driver.get(url1) 
time.sleep(3)
driver.find_element_by_xpath('//*[@id="react-root"]/section/main/article/div[1]/div/div/div[1]/div[1]/a/div[1]/div[2]').click() 

time.sleep(1) 
data = driver.find_element_by_css_selector('.C7I1f.X7jCj') # C7I1f X7jCj 
tag_raw = data.text 
tags = re.findall('#[A-Za-z0-9가-힣]+', tag_raw) 
tag = ''.join(tags).replace("#"," ") # "#" 제거 
tag_data = tag.split() 
for tag_one in tag_data: 
        instagram_tags.append(tag_one) 
print(instagram_tags) 
driver.find_element_by_xpath('/html/body/div[5]/div[1]/div/div/a').click()
time.sleep(3)  
# try : 
for i in range(searchcount): 
    time.sleep(2) 
    data = driver.find_element_by_css_selector('.C7I1f.X7jCj') # C7I1f X7jCj 
    tag_raw = data.text 
    tags = re.findall('#[A-Za-z0-9가-힣]+', tag_raw) 
    tag = ''.join(tags).replace("#"," ") # "#" 제거 
    tag_data = tag.split() 
    for tag_one in tag_data: 
        instagram_tags.append(tag_one) 
    print(tag_one) 
    driver.find_element_by_xpath('/html/body/div[5]/div[1]/div/div/a[2]').click() 
    #driver.close()
# except NoSuchElementException : 
#         driver.close()

#데이터프레임으로 변환 후 파일로 저장
result = pd.DataFrame(instagram_tags)
result.to_csv(filename ,encoding = 'cp949' , index=False)
print('{}로 저장 완료했습니다.' .format(filename))
```
